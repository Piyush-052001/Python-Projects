```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tit = pd.read_csv('titanicc.csv')
tit
```

```python
tit.shape
```

```python
tit.info()
```

```python
tit.describe()
```

```python
# Finding null values
tit.isna().sum()
```

```python
# Deleting the column cabin as it has more null values
tit = tit.drop('Cabin', axis = 1)
```

```python
tit.shape
```

```python
tit['Age'].fillna(np.mean(tit.Age), inplace = True)
```

```python
tit.isna().sum()
```

```python
tit['Embarked'].value_counts()
```

```python
# Replacing null values with mode in embarked column
tit['Embarked'].fillna('S', inplace = True)
```

```python
tit.isna().sum()
```

```python
# Check for duplicates
tit.duplicated().sum()
```

```python
print(f' The average age of Titanic passenger is {np.mean(tit.Age)}')
```

```python
print(f' The average fare of Titanic journey is ${np.mean(tit.Fare)}')
```

```python
tit['Age'].hist()
plt.show()
```

```python
tit.hist('Fare')
plt.show()
```

```python
tit.boxplot('Age')
plt.show()
```

```python
tit.boxplot('Fare')
plt.show()
```

```python
pass_count = tit.Sex.value_counts()
pass_count
```

```python
plt.bar(pass_count.index, pass_count.values)
plt.xlabel('Gender')
plt.ylabel('Passenger count')
plt.title('Sex Representation')
plt.show()
```

```python
# Count the number of passenger in each class
p_class_count = tit.Pclass.value_counts()
```

```python
p_class_count
```

```python
plt.bar(p_class_count.index, p_class_count.values, color = 'red', width = 0.5)
plt.xlabel('Passenger Class')
plt.ylabel('Number of passenger')
plt.show()
```

```python
count_life = tit.Survived.value_counts()
count_life
```

```python
plt.pie(count_life.values, labels = count_life.index, autopct='%1.1f%%', shadow = True, explode = (0.1,0.1), 
        colors =['red','yellow'], 
        wedgeprops = {'edgecolor' : 'black',
                     'linewidth' : 2})
plt.legend()
plt.show()
```

```python
# Finding how many male and female survived
sns.set_style('whitegrid')
sns.countplot(x = 'Survived' , hue ='Sex', data = tit)
plt.show()
```

```python
# Finding how many male and female survived from a particular class
sns.set_style('darkgrid')
sns.countplot(x = 'Survived' , hue ='Pclass', data = tit)
plt.show()
```

```python
plt.scatter(tit.Age, tit.Fare)
plt.xlabel('Age of passenger')
plt.ylabel('Fare of Ticket')
plt.show()
```
