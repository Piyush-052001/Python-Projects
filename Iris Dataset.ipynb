```python
# Box plot or Whiskers plot
sns.boxplot(x = 'species', y = 'petal_length', data  = df)
plt.xlabel('Species of iris flower')
plt.ylabel('Petal length')
plt.show()
```

```python
sns.violinplot(x = 'species' , y = 'petal_width', data = df)
plt.xlabel('Species of Iris flower')
plt.ylabel('PEtal width')
plt.show()
```
