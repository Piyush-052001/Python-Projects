```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
car = pd.read_excel('mtcars.xlsx')
car
```

```python
car.shape
```

```python
car.info()
```

```python
car.describe()
```

```python
car.isna().sum()    # Finding null values
```

```python
# Replacing null values from weight column by mean
car['wt'].fillna(np.mean(car.wt), inplace = True)
```

```python
car.isnull().sum()
```

```python
# Replacing null values from gear column by mode
car['gear'].value_counts()
```

```python
car['gear'].fillna(3.0, inplace = True)
```

```python
car.isnull().sum()
```

```python
# To check the impact of weight on mileage
plt.scatter(x = car.wt, y = car['mpg_high'])
plt.show()
```

```python
# To check the impact of weight on displacement
plt.scatter(x = car.wt, y = car.disp)
plt.xlabel('car weight')
plt.ylabel('Displacement')
plt.show()
```

```python
sns.FacetGrid(car, hue = 'model')\
.map(plt.scatter, 'wt','mpg_city')
plt.show()
```

```python
# Finding number of cylinder in cars
gears = car['am'].value_counts()
```

```python
leb = ['Manual','Automatic']
plt.pie(gears.values, labels=leb)
plt.legend()
plt.show()
```

```python
plt.bar(leb, gears.values)
plt.show()
```

```python
car_cyl = car.cyl.value_counts()
```

```python
plt.bar(car_cyl.index, car_cyl.values, color = 'red')
plt.show()
```

```python
plt.pie(car_cyl.values, labels = car_cyl.index, colors = ['red','yellow','black'])
plt.legend()
plt.show()
```
