```python
# importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ipl = pd.read_csv('matches.csv')
ipl
```

```python
# Display the first few rows 
ipl.head()
```

```python
# Display last few rows
ipl.tail()
```

```python
# Get summary statistics of numerical variables
ipl.describe()
```

```python
# Check the datatypes of variables
ipl.dtypes
```

```python
# Check for null values
ipl.isna().sum()
```

```python
# Filling missing values using mode
ipl['city'].fillna('Mumbai',inplace = True)
```

```python
ipl['winner'].fillna('Mumbai Indians',inplace=True)
```

```python
ipl['player_of_match'].fillna('CH Gayle',inplace=True)
```

```python
ipl['umpire1'].fillna('HDPK Dharmasena',inplace=True)
```

```python
ipl['umpire2'].fillna('SJA Taufel',inplace=True)
```

```python
# Drop the umpire3 column as it has more null values
ipl.drop('umpire3',axis=1,inplace=True)
```

```python
# getting the most man of the match awards
ipl['player_of_match'].value_counts()
```

```python
# finding number of times each team won the toss
ipl['toss_winner'].value_counts()
```

```python
# looking at number of matches played each season
ipl['season'].value_counts()
```

```python
# Finding out how many times a team won the match after winning the toss
np.sum(ipl['toss_winner'] == ipl['winner'])
```

```python
# Top 5 players with most Mom's
ipl['player_of_match'].value_counts()[:5].plot(kind ='barh')
plt.xlabel('Count')
plt.ylabel('Players')
plt.title('Player of the Match')
plt.show()
```

```python
# Top 10 players with most Mom's
plt.figure(figsize=(15,7))
plt.bar((ipl['player_of_match'].value_counts().index[:10]), (ipl['player_of_match'].value_counts()[:10]), color = 'maroon')
plt.xlabel('Players')
plt.ylabel('Count')
plt.title('Player of the Match')
plt.show()
```

Chris Gayle won most Man of the Match Awards.

```python
# Getting frequencies of result column
ipl['result'].value_counts()
```

```python
plt.bar((ipl['result'].value_counts().index), ipl['result'].value_counts(), color='green')
plt.show()
```

There were more Normal Result compared with the other two.

```python
# Number of matches played in each season
plt.figure(figsize=(15,7))
sns.countplot(x='season',
              data=ipl,
              order = ipl['season'].value_counts().index.sort_values()
              )
plt.xlabel('Season')
plt.ylabel('No of matches')
plt.title('Number of matches played in each IPL season')
plt.show()
```

Most number of matches were played in IPL-2013 Season.

```python
# Matches won by the teams
plt.figure(figsize=(15,7))
sns.countplot(y='winner',
              data=ipl,
              order = ipl['winner'].value_counts().index.sort_values()
              )
plt.xlabel('Team')
plt.ylabel('No. of wins')
plt.title('Matches won by the Teams')
plt.show()
```

Mumbai Indians won most number of matches followed by Chennai Super Kings.

```python
# Number of matches played in stadium, Top 10
plt.figure(figsize = (12,8))
sns.countplot(y = 'venue',
              data = ipl,
              order = ipl['venue'].value_counts().iloc[:10].index,
              palette='gist_heat')
plt.xlabel('No of matches')
plt.ylabel('Venue')
plt.title('Number of matches played in each Stadium')
plt.show()
```

Most number of ipl matches were played in M Chinnaswamy Stadium, Banglore.

```python
toss_dec = ipl.toss_decision.value_counts()
toss_dec
```

```python
plt.pie(toss_dec.values, labels = toss_dec.index, autopct='%1.1f%%', shadow=True, explode=(0.1,0.1),
       colors = ['red','pink'],
       wedgeprops={'edgecolor':'black',
                  'linewidth': 2})
plt.title('Teams to choose Bat or Field in Percentage')
plt.legend()
plt.show()
```

In more than 57% of matches, teams chose to field after winning the toss.

```python
win = ipl.winner.value_counts()
win
```

```python
plt.pie(win.values[:8], labels=win.index[:8], autopct='%1.1f%%')
plt.show()
```

The winning percentage of Mumbai Indians is higher followed by Chennai Super Kings.

```python
plt.figure(figsize=(15,7))
sns.countplot(x='season',
              hue='toss_decision',
              data=ipl,
              order = ipl['season'].value_counts().index.sort_values(),
              palette='deep')
plt.xlabel('Season',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.title('Decision to field or bat in each IPL season', fontsize=16)
plt.show()
```

We can see a trend that, decisions to field are more from IPL-2016 Season
